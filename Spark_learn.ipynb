```python
def select(s,e,one_possible):
            if s <= e:
                one_possible.append(s)
                select(s+1,e,one_possible)
                del one_possible[-1]
                select(s+1,e,one_possible)
            result.append(one_possible)
```

```python
def select2(s,e,one_possible):
            if s <= e:
                select2(s+1,e,one_possible + [s])
                select2(s+1,e,one_possible)
            else:
                if one_possible != []:
                    result.append(one_possible)
```
